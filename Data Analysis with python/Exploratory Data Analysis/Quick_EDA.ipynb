### Data Prep

```python
import seaborn as sns
from dataprep.eda import create_report

# Load the Titanic dataset from Seaborn
titanic_data = sns.load_dataset("titanic")

# Generate EDA report
report = create_report(titanic_data)

# Display the report
report.show_browser()
```
